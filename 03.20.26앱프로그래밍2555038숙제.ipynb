```python
!pip install faker

from faker import Faker
import random

fake = Faker()
```

```python
class User:
    user_count = 0

    def __init__(self):
        User.user_count += 1
        self.user_id = User.user_count
        self.name = fake.name()
        self.age = random.randint(18, 60)
        self.address = fake.address().replace("\n", ", ")
        self.email = fake.email()
        self.phone = fake.phone_number()

        self.following = 0
        self.follower = 0

    def following_mode(self, other_user):
        if self.user_id != other_user.user_id:
            self.following += 1
            other_user.follower += 1
            print(
                f"{self.name} → {other_user.name} | "
                f"following: {self.following}, "
                f"follower: {other_user.follower}"
            )

    def __str__(self):
        return (
            f"ID: {self.user_id}, Name: {self.name}, Age: {self.age}, "
            f"Following: {self.following}, Follower: {self.follower}"
        )
```

```python
users = [User() for _ in range(1000)]

print("===== First 5 Users =====")
for user in users[:5]:
    print(user)
```

```python
tom = User()
jerry = User()

tom.name = "Tom"
jerry.name = "Jerry"

print("===== Before =====")
print(tom)
print(jerry)
```

```python
for i in range(10):
    print(f"\n[{i+1} time]")
    tom.following_mode(jerry)
```

```python
print("\n===== After =====")
print(tom)
print(jerry)
```
