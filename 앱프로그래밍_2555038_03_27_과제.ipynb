# JSON 데이터 → DataFrame + address 펼치기

```python
import requests
import pandas as pd

url = "https://jsonplaceholder.typicode.com/users"
data = requests.get(url).json()

df = pd.DataFrame(data)

# address 필드 펼치기
address_df = pd.json_normalize(df['address'])

print(address_df)
```

# 짝수 인덱스 데이터 저장 (Excel)

```python
url = "https://jsonplaceholder.typicode.com/posts"
data = requests.get(url).json()

df = pd.DataFrame(data)

df_even = df[df.index % 2 == 0]

df_even.to_excel("even_posts.xlsx", index=False)
```

# Gradio 인터페이스 (사용자 ID로 게시글 조회)

```python
import gradio as gr
import requests
import pandas as pd

def get_posts(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {"userId": user_id}
    data = requests.get(url, params=params).json()
    return pd.DataFrame(data)

iface = gr.Interface(fn=get_posts, inputs="number", outputs="dataframe")
iface.launch()
```
